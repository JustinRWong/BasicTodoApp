We can see that the corresponding vector has only 2 details all time for each of the tones that it has identified. There are limited amount of tones possible here:

To make the vector, stack all the types of tones in the vector

tone_vector = $\vec{v} = [Anger, Fear, Joy, Sadness, Analytical, Confident, Tentative]^T$

```python
map_to_indexes = {
    "anger": 0,
    "fear": 1,
    "joy": 2,
    "sadness": 3,
    "analytical": 4,
    "confident": 5,
    "tentative": 6
}
def make_tone_vector(tones):
    tones = tones[0]["document_tone"]["tones"]
    result = [0] * 7 #because there are 7 emotions allowed
    for tone in tones:
        result[map_to_indexes[tone["tone_id"]]] = tone["score"]
    return result

print(make_tone_vector(tone_analyzer))
```
